# BAR PLOT LEARNING

```python
import matplotlib.pyplot as plt
```

```python
x = ['python' , 'java' , 'c' , 'reactjs' , 'c++']
y = [90 , 40 , 87 , 35 , 50]
plt.xlabel("languge" , fontsize = 20)
plt.ylabel("range" , fontsize = 20)
plt.title("programming label" , fontsize = 20)
chart_color = ['r','y','m','b','g']
plt.bar(x , y , width = 0.5 , color = chart_color , edgecolor = 'black', linewidth = 1 , linestyle = ':', alpha = 1,label='code varsion')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = ['python' , 'java' , 'c' , 'reactjs' , 'c++']
y = [90 , 40 , 87 , 35 , 50]
z = [22 , 30 , 89 , 65 , 43]

width = 0.5
p = np.arange(len(x))
p1 = [j+width for j in p]

plt.xlabel("languge" , fontsize = 20)
plt.ylabel("range" , fontsize = 20)
plt.title("programming label" , fontsize = 20)

plt.bar(p , y , width , color = 'r', linewidth = 1 ,label='code varsion')
plt.bar(p1 , z , width , color = 'b', linewidth = 1 ,label='code mode')
plt.xticks(p+width/2 , x)
plt.legend()
plt.show()
```

# scatter plot creating

```python
import matplotlib.pyplot as plt
import numpy as np
car = [4,5,6,4,5,6,7,8,9]
sale = [5,4,3,3,2,9,4,5,8]
car_colors = ['red','yellow','pink','red','yellow','pink','black','blue','green']
plt.scatter(car , sale, color=car_colors)
plt.title('scatter plot', fontsize=20)
plt.xlabel("car...", fontsize=15)
plt.ylabel("sale..." ,fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
car = [4,5,6,4,5,6,7,8,9]
sale = [5,4,3,3,2,9,4,5,8]
car_colors = ['red','yellow','pink','red','yellow','pink','black','blue','green']
car_sizes = [300,400,200,500,600,300,700,600,800]
plt.scatter(car , sale, color=car_colors, sizes=car_sizes, marker='*', edgecolor='black', linewidth=1)
plt.title('scatter plot', fontsize=20)
plt.xlabel("car...", fontsize=15)
plt.ylabel("sale..." ,fontsize=15)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
car = [4,5,6,4,5,6,7,8,9]
sale = [5,4,3,3,2,9,4,5,8]
sale2 = [4,5,6,7,8,9,7,6,5]
car_colors = ['red','yellow','pink','red','yellow','pink','black','blue','green']
car_sizes = [300,400,200,500,600,300,700,600,800]
plt.scatter(car , sale, color=car_colors, sizes=car_sizes)
plt.scatter(car , sale2, color='black', sizes=car_sizes)
t = plt.colorbar()
t.set_label('colorbar')
plt.title('scatter plot', fontsize=20)
plt.xlabel("car...", fontsize=15)
plt.ylabel("sale..." ,fontsize=15)
plt.show()
```

# pie chart plot creating

```python
import matplotlib.pyplot as plt
import numpy as np
x = [10,20,40,90,33,50]
y = ['germany','nepal','india','saudi arbia','newzland','france']
plt.pie(x ,labels=y)
plt.show()
```

```python
x = [10,20,40,90,33,50]
y = ['germany','nepal','india','saudi arbia','newzland','france']
ex = [0.0,0.2,0.0,0.1,0.0,0.0]
plt.pie(x ,labels=y,explode=ex,autopct="%0.0f%%",shadow=True,radius=1,labeldistance=1.1)
plt.title("country resio")
plt.legend(loc=2)
plt.show()
```

```python
x = [10,20,40,90,33,50]
y = ['germany','nepal','india','saudi arbia','newzland','france']
ex = [0.0,0.2,0.0,0.1,0.0,0.0]
plt.pie(x ,labels=y,autopct="%0.0f%%",shadow=True,radius=1.5,labeldistance=1.1,startangle=90, textprops={"fontsize":15},wedgeprops={"linewidth":4,"edgecolor":'y'},
                                                                                                                                   rotatelabels=True)
plt.show()
```

```python
x = [10,20,40,90,33,50]
x1 = [30,40,53,60,20,10]
y = ['germany','nepal','india','saudi arbia','newzland','france']
plt.pie(x ,labels=y,radius=1.5)
plt.pie(x1 ,labels=y,radius=0.5)
plt.show()
```

# stem plot creating

```python
import matplotlib.pyplot as plt
import numpy as np
x = [2,3,4,4,3,6,8]
y = [3,2,4,9,5,4,3]
plt.stem(x,y,linefmt=':',bottom=1,basefmt='black',label='temprature',orientation='horizontal')
plt.legend()
plt.show()
```

# stack plot creating

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5,6]
area = [3,4,1,6,3,2]
area2 = [1,3,2,4,5,3]
area1 = [2,3,4,3,1,2]
plt.title("land saler")
plt.xlabel("x-axis")
plt.ylabel('y-axis')
#plt.grid()
plt.stackplot(x,area,area1,area2,labels={'low','high','min'},colors=['r','g','orange'])
plt.legend()
plt.show()
```

# step plot creating

```python
import matplotlib.pyplot as plt
import numpy as np
x = [1,2,3,4,5,5,5,6,8,7,6,7,8,9,9]
y = [4,3,2,3,5,6,3,8,5,6,3,4,5,9,1]
plt.title("step plot matplotlib")
plt.xlabel("x-section")
plt.ylabel("y-section")
plt.step(x,y,color='g',marker='o',ms=4,mfc='r',label='code section time')
plt.legend()
plt.grid()
plt.show()
```
