```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "i am not just human learning its artificial intelligence"
wordcloud = WordCloud().generate(text)
plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

text = "im prepar in human visulization and robotics"
wordcloud = WordCloud(collocations=False,background_color='white',width=2048,height=1080,
                     colormap='viridis').generate(text)

plt.figure(figsize=(10,5))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("word cloud example",fontsize=20)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from PIL import Image
from wordcloud import WordCloud, ImageColorGenerator
import numpy as np

image_path = "C:\\Users\\HARSH\\OneDrive\\Desktop\\ADV_data visulization\\circle.png"
image = Image.open(image_path).convert("RGB")
text = "i am artificial intelligence engineer is proud to be inpiration nsjdjdjd djdjddjjd djfjdjdndjjdddnndnd djdnjddjd nnncfnfdjfdnff fnffnfnffnfn fnfnnfnfn"
mask = np.array(image)
wordcloud = WordCloud(background_color="black",mask=mask).generate(text)

plt.figure(figsize=(4,5))
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = {'category A': 50,
       'category B' : 30,
       'categary C' : 20
       }
fig = plt.figure(FigureClass=Waffle,rows=9,values=data)
plt.show()
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = {
    'car':5,
    'truck':3,
    'plane':2
}

fig = plt.figure(FigureClass=Waffle,rows=4,columns=10,values=data,title={
    'label':'waffle chart vihacal',
    'loc':'center'
},
                legend ={
                    'loc':'upper left',
                    'bbox_to_anchor':(1,1)
                })
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle

data = {'phone':['nokya','motororlo','vivo','oppo','iphone'],
        'stock':[44,5,65,43,23]
       }

df = pd.DataFrame(data)

fig = plt.figure(FigureClass=Waffle,rows=5,values=df.stock,labels=list(df.phone))
plt.show()
            
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle

fig = plt.figure(FigureClass=Waffle,rows=3,columns=3,values=[1,1,1],colors=['red','green','yellow'],
                icons='car',icon_size=5,font_size=20,icon_legend=True,
                legend={
                    'labels':['sun','moon','snow'],
                    'loc':'upper left',
                    'bbox_to_anchor':(1,1)
                })
plt.show()
```
