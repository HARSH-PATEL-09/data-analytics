##Waffle Chart

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
data = {'category A': 50,
       'category B': 30,
       'category c': 20
       }
fig = plt.figure(FigureClass=Waffle, rows=10, values=data)
plt.show()
```

```python
data = {
    'step 1' : 5,
    'step 2' : 3,
    'step 3' : 2
}
fig1 = plt.figure(FigureClass = Waffle , rows = 10, values = data)
fig2 = plt.figure(FigureClass = Waffle , columns = 10 , values = data)
fig3 = plt.figure(FigureClass = Waffle , rows = 10 , columns = 10 , values = data)
plt.show()
```

```python
dataset = {
    'normal' : 50,
    'highlevel' : 30,
    'dicent' : 20
}
fig = plt.figure(FigureClass = Waffle , rows = 5 ,values = dataset,
                title = {'label':'waffle chart example','loc':'center'},
                legend = {'loc':'upper left',
                         'bbox_to_anchor':(1,1)})
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from pywaffle import Waffle
phone_data = {'phone' : ['sumsang','apple','nokia','blackbary'],
             'stock' : [44,43,21,10]
             }
df = pd.DataFrame(phone_data)
phone_fig = plt.figure(FigureClass = Waffle , rows = 5 , values = df.stock, labels = list(df.phone))
plt.show()
```

```python
import matplotlib.pyplot as plt
from pywaffle import Waffle
gander_data = {'men': 60,
              'women': 40
              }
fig = plt.figure(FigureClass = Waffle, 
                rows = 10,
                values = gander_data,
                colors = ["g","r"],
                icons = 'user',
                icon_size = 1,
                title = {'label':'gander distribution',
                        'loc':'center'}
                )
plt.show()
```

```python
from pywaffle import Waffle
import matplotlib.pyplot as plt
data = {'Men': 6, 'Women': 4}
fig = plt.figure(
FigureClass=Waffle,
rows=10,
values=data,
colors=["#4CAF50", "#FFC107"],
icons='user',
icon_size=2,
title={'label': 'Gender Distribution', 'loc': 'center'}
)
plt.show()
```
