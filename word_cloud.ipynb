```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
text = "my name is harsh kothiya i belive in god"
wordcloud = WordCloud().generate(text)
plt.figure(figsize=(9, 9))
plt.imshow(wordcloud)
plt.show()
```
