```python
import numpy as np
import matplotlib.pyplot as plt
Name = ("Aakash", "Shalini", "Rahul", "Raj")
Marks = {
'ADA': (25,20,26,18),
'PDS': (27,28,20,17),
'CN': (20,21,22,26),
'SE':(20,12,18,24)
}
x = np.arange(len(Name))
plt.bar(x, Marks['ADA'], 0.1, color='red', label = 'ADA')
plt.bar(x + 0.1, Marks['PDS'], 0.1, color='orange', label = 'PDS')
plt.bar(x + 0.2, Marks['CN'], 0.1, color='yellow', label = 'CN')
plt.bar(x + 0.3, Marks['SE'], 0.1, color='gray', label = 'SE')
plt.xticks(x, Name)
plt.xlabel("Student Name")
plt.ylabel("Subject wise Marks")
plt.title("Student Result")
plt.legend(loc='best',ncol=4)
plt.ylim(0, 40)
plt.show()
```
