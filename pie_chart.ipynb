###pie chart

```python
import matplotlib.pyplot as plt

categary = ['nepal','japan','malacia','india','usa','soudi arbian']
value = [20,40,30,59,60,70]

plt.pie(value,labels=categary)
plt.show()
```

```python
import matplotlib.pyplot as plt

name = ['farari','mastangeGT','i20','audi','lambognisentariyo','creta','siaze','fanty']
stock = [43,56,78,43,23,79,90,54]
most_priority = [0,0,0,0,0,0,0.3,0]

plt.pie(stock,labels=name,
        startangle=90,
        counterclock=False,
        explode=most_priority,
       shadow=True
        
       )


plt.show()
```

```python
import matplotlib.pyplot as plt

hours = list(range(1,13))
values = [1]*12

plt.pie(values,labels=hours,startangle=90,counterclock=False)
plt.show()
```

```python
import matplotlib.pyplot as plt
import datetime as dt

print(dt.datetime.now())
hours = list(range(1,13))
curr_hour = dt.datetime.now().hour%12
slice=[0]*12

for i in range(0,12):
  if(i==curr_hour):
      slice[1-i]=0.2
      break

      print(slice)

valuse = [1]*12
plt.pie(valuse,labels=hours,startangle=75,counterclock=False,explode=slice)
plt.show()
```
